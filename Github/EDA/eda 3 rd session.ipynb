# Changing the Plot Styple with set_style

# Set plot background to a white grid

```python
sns.set()
sns.set_style("whitegrid")
sns.lmplot(x='beer_servings',y='wine_servings',data=df);
```

```python
sns.set()
sns.set_style("darkgrid")
sns.lmplot(x='beer_servings',y='wine_servings',data=df);
```

```python
sns.set()
sns.set_style("white")
sns.lmplot(x='beer_servings',y='wine_servings',data=df);
```

```python
sns.set()
sns.set_style("whitegrid")
sns.swarmplot (x='country',y='wine_servings',data=df);
```

```python
sns.set()
sns.set_style("dark")
sns.lmplot(x='beer_servings',y='wine_servings',data=df);
```

```python
sns.set()
sns.set_style("whitegrid")
sns.swarmplot (x='country',y='wine_servings',data=df);
```

```python
sns.set()
sns.set_style("ticks", {"axes.facecolor":".1"})
sns.boxplot(data=df)
```

```python
sns.set()
sns.set_context("talk")
plt.figure(figsize=(8,6))
sns.lmplot(x='beer_servings', y='wine_servings', data=df)
```
