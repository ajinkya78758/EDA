# MLB Players Heigh, Weight, Age and Positions dataset

```python
mlb = pd.read_csv("data-mlb-players.csv")
```

# Plot it with PairGrid()

```python
g = sns.PairGrid(mlb, vars=["Height", "Weight"], hue="Position")
g.map(plt.scatter);
g.add_legend();
```

# Plotting with PairPlot()
